```python
def genChart(data, title):
    
    fig, ax = plt.subplots(figsize=(10,6))
    
    x = np.arange(len(data.name.unique()))
    
    bar_width = 0.4
    
    b1 = ax.bar(x, data['throughput'], width=bar_width, label='throughput', color="cornflowerblue")
    ax.set_ylabel("Throughput (req/s)",color="cornflowerblue",fontsize=14)
    
    ax2=ax.twinx()
    b2 = ax2.bar(x+bar_width, data['latency'], width=bar_width, label='latency', color="lightcoral")
    ax2.set_ylabel("Latency (ms/req)",color="lightcoral",fontsize=14)
    
    ax.set_xticks(x + bar_width / 2)
    ax.set_xticklabels(data.name.unique())

    # Add legend.


    # Axis styling.
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('#DDDDDD')
    ax.tick_params(bottom=False, left=False)
    ax.set_axisbelow(True)
    ax.yaxis.grid(True, color='#EEEEEE')
    ax.xaxis.grid(False)

    # Add axis and chart labels.
    ax.set_xlabel('Actors', labelpad=15)
    ##ax.set_title('Performance during 20000 activations of PatientGrain and UserGrain', pad=15)

    fig.tight_layout()
    fig.savefig(title+'.eps',format='eps', dpi=1000, bbox_inches='tight')
```

```python
def genChart2(data, title):
    fig, ax = plt.subplots(figsize=(10,6))
    
    x = np.arange(len(data.name.unique()))
    bar_width = 0.4
    
    ax.bar(x, data['total']*1000, width=bar_width, label='Total time', color="cornflowerblue")
    
    ax.set_ylabel("Total time (ms)",color="cornflowerblue",fontsize=14)
    ax.set_xticks(x)
    ax.set_xticklabels(data.name.unique())
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('#DDDDDD')
    ax.tick_params(bottom=False, left=False)
    ax.set_axisbelow(True)
    ax.yaxis.grid(True, color='#EEEEEE')
    ax.xaxis.grid(False)

    # Add axis and chart labels.
    ax.set_xlabel('Methods', labelpad=15)
    ##ax.set_title('Performance during 20000 activations of PatientGrain and UserGrain', pad=15)

    fig.tight_layout()
    fig.savefig(title+'.eps',format='eps', dpi=1000, bbox_inches='tight')
```

```python
def genLineChart(data, x, y_label, title):
    fig, ax = plt.subplots(figsize=(10,6))
    
    lines = data.name.unique()

    for name in lines:
        y = data[data['name'] == name]['values'].tolist()[0]
        ax.plot(x, y, marker='o', label=name)
    
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('#DDDDDD')
    
    ax.tick_params(bottom=False, left=False)
    ax.set_axisbelow(True)
    ax.yaxis.grid(True, color='#EEEEEE')
    ax.xaxis.grid(False)

    # Add axis and chart labels.
    ax.legend()
    ax.set_xlabel('Number of activations', labelpad=15)
    ax.set_ylabel(y_label, labelpad=15)
    ##ax.set_title('Performance during 20000 activations of PatientGrain and UserGrain', pad=15)

    fig.tight_layout()
    fig.savefig(title+'.eps',format='eps', dpi=1000, bbox_inches='tight')
```
